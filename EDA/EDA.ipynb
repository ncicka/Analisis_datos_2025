# **Visualización**

```python
# Estilo general de los graficos
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (14,6)
```

```python
#Toneladas cosechadas por provincia
toneladas_cosechadas_provincia = df.groupby(['provincia_nombre'])['produccion_tm'].sum().reset_index(name='cantidad_toneladas')
toneladas_cosechadas_provincia.sort_values(['cantidad_toneladas','provincia_nombre'], ascending=True, inplace=True)
toneladas_cosechadas_provincia
```

```python
#LinePlot: Produccion en tn de las ultimas 25 campañas, de las cinco provincias de mayor produccion

# Ordenar por Campaña
df_sorted = df.sort_values('campania')

# Cuatro provincias con mayor produccion exceptuando Buenos Aires
provincia_filtro = ['Chaco','Córdoba','Santa Fe','La Pampa','Buenos Aires']

# ultimas 26 campañas
campania_filtro = ['1995/96','1996/97','1997/98','1990/00','2000/01','2001/02','2002/03','2003/04','2004/05','2005/06','2006/07','2007/08','2009/10',
                   '2010/11','2011/12','2012/13','2013/14','2014/15','2015/16','2016/17','2017/18','2019/20','2020/21','2021/22','2022/23','2023/24']

df_filtrado = df_sorted[df_sorted['provincia_nombre'].isin(provincia_filtro)]

df_filtrado = df_filtrado[df_filtrado['campania'].isin(campania_filtro)]

plt.figure(figsize=(12,6))

# Grafico de lineas
sns.lineplot(data=df_filtrado, x='campania', y='produccion_tm', hue='provincia_nombre')
plt.title('Tendencia de produccion de Girasol por campaña')
plt.xlabel('Campaña')
plt.ylabel('Produccion anual')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# BoxPlot: Distribucion de la produccion por provincias

sns.boxplot(data=df, x='provincia_nombre', y='produccion_tm')
plt.title('Distribucion de produccion')
plt.xlabel('Campaña')
plt.ylabel('Tn')
plt.xticks(rotation=90)
plt.show()
```

```python
# Histograma: Superficie sembrada en has

sns.histplot(data=df, x='superficie_sembrada_ha', bins=50, kde=True)
plt.title('Superficie sembrada Has')
plt.xlabel('Has sembradas')
plt.ylabel('Frecuencia')
plt.show()
```

```python
#Produccion en toneladas por departamento y campaña, de la provincia del Chaco
provincia_filtro = ['Chaco']
campania_filtro = ['2021/22','2022/23','2023/24']

df_filtrado = df[df['provincia_nombre'].isin(provincia_filtro)]

df_filtrado = df_filtrado[df_filtrado['campania'].isin(campania_filtro)]

sns.barplot(data=df_filtrado,
            x='departamento_nombre', y='produccion_tm',
             hue='campania',errorbar=None)


plt.xticks(rotation=90)
plt.title('Total produccion de Girasol por departamento Prov. Chaco')
plt.xlabel('Departamento')
plt.ylabel('Toneladas')
plt.show()
```

```python
#BarPlot: Total de produccion por campaña en la provincia del Chaco

provincia_filtro = ['Chaco']
campania_filtro = ['2020/21','2021/22','2022/23','2023/24']
departamento_filtro = ['Comandante Fernández']

df_filtrado = df[df['provincia_nombre'].isin(provincia_filtro)]

df_filtrado = df_filtrado[df_filtrado['campania'].isin(campania_filtro)]

df_filtrado = df_filtrado[df_filtrado['departamento_nombre'].isin(departamento_filtro)]


sns.barplot(data=df_filtrado,
            x='campania', y='produccion_tm',
             errorbar=None)

plt.xticks(rotation=90)
plt.title('Total produccion Girasol en Chaco')
plt.xlabel('Campaña')
plt.ylabel('Toneladas')
plt.show()
```

```python
# Pivot para matriz de calor - con filtro (dos al mismo tiempo)

provincia_pivot = ['Chaco']

campania_pivot = ['2020/21','2021/22','2022/23','2023/24']

df_filtrado = df[df['provincia_nombre'].isin(provincia_pivot)]

df_filtrado = df_filtrado[df_filtrado['campania'].isin(campania_pivot)]


# Check if the filtered DataFrame is empty
if df_filtrado.empty:
    print("No data found for the selected filters.")
else:
    pivot = df_filtrado.pivot_table(index='departamento_nombre', columns='campania', values='produccion_tm', aggfunc='sum')

    # Graficar
    plt.figure(figsize=(10, 8))
    sns.heatmap(pivot, cmap="YlOrRd", linewidths=0.3)
    plt.title("Produccion en Tn en la provincia del chaco")
    plt.xlabel("Campaña")
    plt.ylabel("Departamentos")
    plt.show()
```
